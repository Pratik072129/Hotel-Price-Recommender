```python
# prompt: all method avg and make 5 differnet plot

import matplotlib.pyplot as plt

# Assuming accuracy_results dictionary is already populated

# Calculate the average accuracy
average_accuracy = sum(accuracy_results.values()) / len(accuracy_results)
print(f"\nAverage accuracy across all models: {average_accuracy:.2f}%")


# 1. Bar plot of model accuracies
plt.figure(figsize=(10, 6))
plt.bar(accuracy_results.keys(), accuracy_results.values())
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.title("Model Accuracies")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# 2. Horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(list(accuracy_results.keys()), list(accuracy_results.values()))
plt.xlabel("Accuracy (%)")
plt.ylabel("Models")
plt.title("Model Accuracies (Horizontal)")
plt.show()


# 3. Scatter plot (Model vs. Accuracy)
plt.figure(figsize=(10, 6))
plt.scatter(accuracy_results.keys(), accuracy_results.values())
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.title("Model Accuracies (Scatter)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


# 4. Pie chart of model accuracies
plt.figure(figsize=(8, 8))
plt.pie(accuracy_results.values(), labels=accuracy_results.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Model Accuracies (Pie Chart)")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# 5. Line plot (Model vs. Accuracy) -  Less suitable but demonstrating different plot type
plt.figure(figsize=(10, 6))
plt.plot(list(accuracy_results.keys()), list(accuracy_results.values()), marker='o')
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.title("Model Accuracies (Line Plot)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
